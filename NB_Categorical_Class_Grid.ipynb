```python
import pandas as pd
dataset = pd.read_csv("CKD.csv")
dataset = pd.get_dummies(dataset, drop_first=True, dtype=int)
dataset
```

```python
dataset.columns
```

```python
ind = dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dep = dataset[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(ind, dep, test_size=0.3, random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import CategoricalNB
param_grid = {'alpha' : [1.0, 2.0],
              'min_categories' : [2, 3]}

grid = GridSearchCV(CategoricalNB(), param_grid, refit = True, verbose =3 , n_jobs=-1, scoring='f1_weighted')
grid.fit(X_train, y_train)
```

```python
re = grid.cv_results_
table = pd.DataFrame.from_dict(re)
table.to_excel("CategoricalNB.xlsx")
table
```

```python
grid_pred = grid.predict(X_test)
```
